```python
import numpy as np
import matplotlib.pyplot as plt 

n = 100000
U = np.random.uniform(0,1,size=n)

X1 = 1-np.sqrt(U)
X2 = np.sqrt(U)

lbd = 1
X3 = -np.log(U)/(3*lbd)

print('(a) E[X]: ', np.mean(X1))
print('(b) E[X]: ', np.mean(X2))
print('(c) E[X]: ', np.mean(X3))
```

##Q2

```python
X = np.ones(10000)
for i in range(10000):
  j = 0
  lbd = 1
  p = np.exp(-lbd)
  F = p
  U = np.random.uniform(0,1)
  while U > F:
    p = lbd*p/(j+1)
    F += p
    j += 1
  X[i] = j

print('mean: ', np.mean(X))
print('variance: ', np.var(X))
```

##Q4
set g(x) = 1<br>since 2(1-Y)<=a for all Y in [0,1], set a=2

```python
n, accept, M = 0, 0, 100000
X = []
a = 2
while accept < M:
	Y = np.random.uniform(0,1)
	U = np.random.uniform(0,1)
	if a*U <= 2*(1-Y) :
		X.append(Y)
		accept += 1
	n += 1

print('the average number of iterations: ', n/accept)
```
